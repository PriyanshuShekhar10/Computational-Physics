```python
# (a)

import math

def func(x):
    return x**x + x - 4

def derivative(x):
    return x**x * (1 + 2 * (math.log(x)))

x_0 = 1.5
tolerance = 1e-6
max_iterations = 10

for i in range(max_iterations):
    f_x_0 = func(x_0)
    f_prime_x_0 = derivative(x_0)
    x_1 = x_0 - f_x_0 / f_prime_x_0

    if abs(x_1 - x_0) < tolerance:
        break

    x_0 = x_1

print("Approximate root:", x_1)
```
