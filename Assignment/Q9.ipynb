```python
# (a) part

import matplotlib.pyplot as plt

def plot(b):
    a = 10
    c = 8/3

    def f(x, y, z):
        x_bar = a * (y - x)
        y_bar = b * x - y - x * z
        z_bar = x * y - c * z
        return x_bar, y_bar, z_bar

    x0, y0, z0, dt = 1, 0, 0, 0.0001
    xval, yval, zval, tval = [x0], [y0], [z0], [0]

    for _ in range(int(50/dt)):
        tval.append(tval[-1] + dt)


        k1x, k1y, k1z = f(x0, y0, z0)
        k2x, k2y, k2z = f(x0 + 0.5 * dt * k1x, y0 + 0.5 * dt * k1y, z0 + 0.5 * dt * k1z)
        k3x, k3y, k3z = f(x0 + 0.5 * dt * k2x, y0 + 0.5 * dt * k2y, z0 + 0.5 * dt * k2z)
        k4x, k4y, k4z = f(x0 + dt * k3x, y0 + dt * k3y, z0 + dt * k3z)

        x1 = x0 + (dt / 6) * (k1x + 2*k2x + 2*k3x + k4x)
        y1 = y0 + (dt / 6) * (k1y + 2*k2y + 2*k3y + k4y)
        z1 = z0 + (dt / 6) * (k1z + 2*k2z + 2*k3z + k4z)

        xval.append(x1)
        yval.append(y1)
        zval.append(z1)
        x0, y0, z0 = x1, y1, z1

    plt.plot(zval, tval, label=f'b={b}')

# Plot for different values of b
plot(5)
plot(10)
plot(25)

plt.xlabel('z')
plt.ylabel('t')
plt.legend()
plt.show()
```
