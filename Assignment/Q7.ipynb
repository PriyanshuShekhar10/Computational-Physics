```python
# for N0 = 100

import matplotlib.pyplot as plt
import random

N0 = 100
p = 0.01
time = [N0]
updatedNuclei = [N0]
tmax = 100


for i in range(tmax):
  time.append(i+1)
  for _ in range(N0):
    if random.random() < p:
      N0 = N0 - 1
  updatedNuclei.append(N0)

# print(updatedNuclei)
plt.grid(True)
plt.plot(updatedNuclei)
plt.xlabel('time')
plt.ylabel('N')
plt.show()
```
