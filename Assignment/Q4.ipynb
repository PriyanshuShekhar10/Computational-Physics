```python
# Using Monte-Carlo method

import random

def fn(x):
  return (math.exp(-math.cos(x**2)))/(math.sqrt(1 - x**2))

random_numbers = []
f_values = []
a = -1
b = +1
N = 10
sum = 0

for i in range(N):
  r = random.random()*2 - 1
  random_numbers.append(r)

for j in range(N):
  f_values.append(fn(random_numbers[j]))

for _ in range(N):
  sum = sum + f_values[_]

integral = (b-a)*sum/N

print('Integral is: ' ,integral)
```
