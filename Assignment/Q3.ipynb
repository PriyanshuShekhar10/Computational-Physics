```python
# using trapezoidal rule

import math
from matplotlib import pyplot as plt
import numpy as np


def fn(x):
  return (math.log(math.cos(x))) * (math.e) ** (math.sin(x))


b  = math.pi/2 #upperlimit
a = 0          #lowerlimit
N  = 5    #number of points
h = (b - a)/N  #width


sum = 0

sum = 0
for i in range(1, N):
    xi = a + i * h
    sum += fn(xi)

integral = h * (0.5 * fn(a) + sum + 0.5 * fn(b))

print('Integral is: ' ,integral)
```

```python
# using gauss legendre approach

def fn(x):
  return (math.log(math.cos(x))) * (math.e) ** (math.sin(x))

weighted_sum = 0
abscissas = [-0.90617985, -0.53846931,0,  0.5384693,0.90617985]
weights = [0.23692689, 0.47862867, 0.56888889, 0.47862867, 0.23692689 ]
scaled_abscissas = []
f_values = []
b = math.pi/2 #upper limit
a = 0 #lower limit
for i in range(len(abscissas)):
  scaled_abscissas.append(abscissas[i]*(b-a)/2 + (b+a)/2)
# print(scaled_abscissas)

for j in range(len(abscissas)):
  f_values.append(fn(math.radians(math.degrees(scaled_abscissas[j]))))
# print(f_values)
for _ in range(len(weights)):
  weighted_sum = weighted_sum + f_values[_]*weights[_]

integral = ((b-a)/2)*weighted_sum


print('Integral is: ' ,integral)
```
