```python
# Using RK-4 method
import numpy as np
import matplotlib.pyplot as plt

x_0 = 0
y_0 = 1
h = 0.2
x_1 = 1.0

x_val= [x_0]
y_val= [y_0]

while x_val[-1] < x_1:
    x_n = x_val[-1]
    y_n = y_val[-1]

    k1 = h * (x_n * y_n + y_n**2 - 2)
    k2 = h * ((x_n + 0.5 * h) * (y_n + 0.5 * k1) + (y_n + 0.5 * k1)**2 - 2)
    k3 = h * ((x_n + 0.5 * h) * (y_n + 0.5 * k2) + (y_n + 0.5 * k2)**2 - 2)
    k4 = h * ((x_n + h) * (y_n + k3) + (y_n + k3)**2 - 2)

    y_new = y_n + (k1 + 2*k2 + 2*k3 + k4) / 6.0
    x_new = x_val[-1] + h

    x_val.append(x_new)
    y_val.append(y_new)

print("RK-4 method estimate of y(1):", y_val[-1])

plt.plot(x_val, y_val, 'green')
plt.xlabel('x')
plt.ylabel('y')
# plt.legend()
plt.show()
```

```python
# Using euler's method
import numpy as np
import matplotlib.pyplot as plt
x_0 = 0
y_0 = 1
h = 0.2
x_target = 1.0

x_values_euler = [x_0]
y_values_euler = [y_0]

while x_values_euler[-1] < x_target:
    x_new = x_values_euler[-1] + h
    y_new = y_values_euler[-1] + h * (x_values_euler[-1] * y_values_euler[-1] + y_values_euler[-1]**2 - 2)

    x_values_euler.append(x_new)
    y_values_euler.append(y_new)

print("Euler's method estimate of y(1):", y_values_euler[-1])

plt.plot(x_values_euler, y_values_euler,'orange')
plt.xlabel('x')
plt.ylabel('y')
# plt.legend()
plt.show()
```
