```python
# using trapezoidal rule

import math
from matplotlib import pyplot as plt
import numpy as np


def fn(x):
  return 1/(x*math.log(x))


b  = math.e + 1 #upperlimit
a = math.e          #lowerlimit
N  = 5    #number of points
h = (b - a)/N  #width


sum = 0

sum = 0
for i in range(1, N):
    xi = a + i * h
    sum += fn(xi)

integral = h * (0.5 * fn(a) + sum + 0.5 * fn(b))

print('Integral is: ' ,integral)
```
