```python
import numpy as np

m = 1
k = 10
x0 = 1
v0 = 0
t0 = 0
t_final = 1
h = 0.2

num_steps_euler = int((t_final - t0) / h) + 1
t_values_euler = np.linspace(t0, t_final, num_steps_euler)
x_values_euler = np.zeros(num_steps_euler)
v_values_euler = np.zeros(num_steps_euler)

x_values_euler[0] = x0
v_values_euler[0] = v0

for i in range(1, num_steps_euler):
    x_dot, v_dot = v_values_euler[i-1], -k/m * x_values_euler[i-1]
    x_values_euler[i] = x_values_euler[i-1] + h * x_dot
    v_values_euler[i] = v_values_euler[i-1] + h * v_dot

print(f"t(1) = {t_values_euler[-1]}, x(1) = {x_values_euler[-1]}, v(1) = {v_values_euler[-1]}")
```

```python
# Using RK-4

import numpy as np

m = 1
k = 10
x0 = 1
v0 = 0
t0 = 0
t_final = 1
h = 0.2

num_steps_rk4 = int((t_final - t0) / h) + 1
t_values_rk4 = np.linspace(t0, t_final, num_steps_rk4)
x_values_rk4 = np.zeros(num_steps_rk4)
v_values_rk4 = np.zeros(num_steps_rk4)

x_values_rk4[0] = x0
v_values_rk4[0] = v0

for i in range(1, num_steps_rk4):
    k1x, k1v = v_values_rk4[i-1], -k/m * x_values_rk4[i-1]
    k2x, k2v = v_values_rk4[i-1] + h/2 * k1v, -k/m * (x_values_rk4[i-1] + h/2 * k1x)
    k3x, k3v = v_values_rk4[i-1] + h/2 * k2v, -k/m * (x_values_rk4[i-1] + h/2 * k2x)
    k4x, k4v = v_values_rk4[i-1] + h * k3v, -k/m * (x_values_rk4[i-1] + h * k3x)

    x_values_rk4[i] = x_values_rk4[i-1] + h/6 * (k1x + 2*k2x + 2*k3x + k4x)
    v_values_rk4[i] = v_values_rk4[i-1] + h/6 * (k1v + 2*k2v + 2*k3v + k4v)

print(f"t(1) = {t_values_rk4[-1]}, x(1) = {x_values_rk4[-1]}, v(1) = {v_values_rk4[-1]}")
```
