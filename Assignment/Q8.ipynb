```python
# (a) Part
import numpy as np

b = 3
a = 0
def p(x):
    return np.exp(-x)

def fn(x):
    return x**(3/2) * np.exp(-x)

N = 10000

random_values = np.random.exponential(scale=1, size=N) % b

integral = np.mean(fn(random_values) / p(random_values))

print("Integral is : ", integral)
```

```python
# (b) Part

b = np.pi
a = 0
def p(x):
    return np.exp(-x) / (np.exp(-0) - np.exp(-np.pi))

def fn(x):
    return 1 / (x**2 + np.cos(x)**2)

N = 10000

random_values  = np.random.exponential(scale=1, size=N) % b

integral = np.mean(fn(random_values ) / p(random_values ))

print("Integral is : ", integral)
```
